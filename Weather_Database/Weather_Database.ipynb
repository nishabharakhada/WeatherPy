```python
import numpy as np
import pandas as pd
import requests
from citipy import citipy
from config import api_key
```

```python
lt = np.random.uniform(low=-90.00, high = 90.00,size=2000)
lg = np.random.uniform(low=-180.00, high = 180.00,size=2000)
coordinates = list (zip(lt,lg))
coordinates
```

```python
city_list = []
for c in coordinates: 
    city = citipy.nearest_city(c[0],c[1]).city_name
    if city not in city_list:
        city_list.append(city)
city_list
```
